```python
batters = soup.find_all("div", class_="cb-col cb-col-25")
print(batters)
bowlers = soup.find_all("div", class_="cb-col cb-col-38")
print(bowlers)
```

```python
print(len(bowlers))
```

```python
bowler_names = []
for bowler in bowlers:
    name = bowler.find('a', class_='cb-text-link').text.strip()
    bowler_names.append(name)
print(bowler_names)
```

```python
batter_names = []
for batter in batters:
    name = batter.find('a', class_='cb-text-link').text.strip()
    batter_names.append(name)
print(batter_names)
```

```python
print(len(batter_names))
print(len(bowler_names))
```

```python
lines = soup.find_all("div", class_="cb-col cb-col-100 cb-scrd-itms")
print(lines)
print(len(lines))
# bowlers = soup.find_all("div", class_="cb-col cb-col-38")
# print(bowlers)
```

```python
batters_scores = {}
bowlers_wickets = {}
for line in lines:
    isBatter = None
    isBolwer = None
    isBatter = line.find("div", class_="cb-col cb-col-25")
    isBolwer = line.find("div", class_="cb-col cb-col-38")
    if isBatter:
        name = isBatter.find('a', class_='cb-text-link').text.strip()
        score = line.find("div", class_="cb-col cb-col-8 text-right text-bold").text.strip()
        batters_scores[name] = int(score)

    elif isBolwer:
        name = isBolwer.find('a', class_='cb-text-link').text.strip()
        score = line.find("div", class_="cb-col cb-col-8 text-right text-bold").text.strip()
        bowlers_wickets[name] = int(score)
print(batters_scores)
print(bowlers_wickets)
```
